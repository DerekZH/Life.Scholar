```python
import pandas as pd
import json
import requests, urllib
from bs4 import BeautifulSoup
import os
from random import randint
```

```python
#this is ugly code, needing to be better

def combine_list_of_lists(lol):
    text = ''
    for i in lol:
        try:
            for j in i:
                try:
                    text = text + j.contents                        
                except:
                    try:
                        text = text + combine_list_of_lists(j.contents)
                    except:
                        text = text + j
        except:
            pass
    return text
```

```python
with open('Data/courses.json', 'rb') as f:
    courses = json.load(f)
courses = pd.DataFrame(courses['elements'])
```

### Getting all course description pages

```python
for courseName in courses.shortName:
    print courseName
    url = 'https://www.coursera.org/course/%s' %courseName
    r = requests.get(url)
    with open('CourseWebpages/%s.html'%courseName, 'wb') as f:
        f.write(r.text.encode('utf-8'))
```

### Beautiful Souping

```python
all_courses = []
for file_name in os.listdir('CourseWebpages/'):
    try:
        with open('CourseWebpages/%s'%file_name, 'rb') as f:
            html = f.read()

        course_info = {}
        course_info['course_name'] = file_name.strip('.html')
        soup = BeautifulSoup(html)
        a = soup.findAll(attrs={'class':'c-cd-section'})

        #### course description

        b = a[0].findAll('p') + a[0].findAll('div') + a[0].findAll('li') + a[0].findAll('ul') + a[0].findAll('span')

        text_b = ''
        for p in b:
            p = p.contents
            try:
                for item in p:
                    try:
                        text_b = text_b + item.output_ready()
                    except:
                        for content in item.contents:
                            text_b = text_b + contents
            except:
                pass

        course_info['course_des'] = text_b

        #### syllabus

        c = a[1]
        c = c.findAll('p') + c.findAll('li') + c.findAll('div')
        c = [item.contents for item in c]
        c = combine_list_of_lists(c)

        course_info['syllabus'] = c

        #### recommended background

        d = a[2]

        try:
            d = [item.contents for item in d]
            d = combine_list_of_lists(d)
        except:
            text_d = ''
            for item in d.contents:
                try:
                    text_d = text_d + ' ' + item.contents[0]
                except:
                    pass    
            d = text_d

        course_info['recom_bkgd'] = d

        #### FAQ

        e = a[-1]
        e = e.findAll('p') + e.findAll('li') + e.findAll('div')
        e = [item.contents for item in e]
        e = combine_list_of_lists(e)

        course_info['faq'] = e

        all_courses.append(course_info)
    except:
        pass
```

```python
all_courses = pd.DataFrame(all_courses)
```

```python
import nltk
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
 
```

```python

def paragraph_stemmer(string): #and non english word remover
    porter = nltk.stem.porter.PorterStemmer()
    string = string.replace('\n',' ').replace(',',' ').replace('.',' ').replace('!',' ').replace('?',' ').replace('\r',' ').split(' ')
    new_string = []
    removed_foreign_words = []
    for item in string:
        try:
            new_string.append(porter.stem(item.encode('utf-8').lower()))
        except:
            pass
    return ' '.join(new_string), removed_foreign_words
```

```python
bkgd_text = paragraph_stemmer(text)
```
